```python
Accession_ids = ['GU190215.1', 'KJ473815.1', 'GQ153547.1' , 'MN996532.1' , 'KY417150.1' , 'KJ473816.1' , 'MK211376.1' , 'MG772933.1' , 'MG772934.1' , 'AY572034.1' , 'MT019529.1' , 'AY291315.1' , 'MT040334.1' , 'MT072864.1' , 'MT084071.1' , 'LC556375.1', 'MW251311' , 'MW251310.1' , 'MW251309.1' , 'MW251312']
```

```python
evaluated_genes = ['1ab','orf3a','orf6','orf7a','orf8','orf10','spike','envelope','membrane','nucleocapsid']
evaluated_genes_dic = {'1ab':[],'orf3a':[],'orf6':[],'orf7a':[],'orf8':[],'orf10':[],'spike':[],'envelope':[],'membrane':[],'nucleocapsid':[]}
```

```python
lst_record = []
for id in Accession_ids:
  handle = Entrez.efetch(db='nucleotide', id=id, rettype='gb')
  record = SeqIO.read(handle, "genbank")
  handle.close()
  print(id)
  print(record.annotations['source'])
  print("***************")
  lst_record.append(record)

SeqIO.write(lst_record, "/content/drive/My Drive/BioinformaticsProject/Experiment1/cov_sample.fasta", "fasta")
```

```python
for id in Accession_ids:
  handle = Entrez.efetch(db='nucleotide', id=id, rettype='gb')
  record = SeqIO.read(handle, "genbank")
  handle.close()
  for feature in record.features:
    if feature.type == 'CDS':
      gene = feature.qualifiers['product'][0].lower()
      for ev_gene in evaluated_genes:
        if ev_gene in gene or "surface glycoprotein" in gene:
          entry = f">{record.annotations['source']} ({id}) \n"
          entry = entry.replace(" ","-")
          entry += record.seq[feature.location.start : feature.location.end]
          if "surface glycoprotein" in gene:
            evaluated_genes_dic["spike"].append(entry)
          evaluated_genes_dic[ev_gene].append(entry)
          break
```

```python
for key, value in evaluated_genes_dic.items():
  data = ""
  for seq in evaluated_genes_dic[key]:
    data += seq + "\n"
  file = open(f'/content/drive/My Drive/BioinformaticsProject/Experiment1/Homologous_Sequences/{key}.fasta', "w")
  file.write(str(data))
  file.close()
```
